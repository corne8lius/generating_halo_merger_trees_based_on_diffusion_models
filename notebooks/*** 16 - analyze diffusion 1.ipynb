```python
def percentage_values_in_range(tensor, lower_bound, upper_bound):
    nonzero_indices = torch.nonzero(tensor.flatten())
    nonzero_values = tensor.flatten()[nonzero_indices[:, 0]]
    # Count the number of values in the range
    num_values_in_range = torch.sum((nonzero_values >= lower_bound) & (nonzero_values <= upper_bound))
    
    # Calculate the total number of values in the tensor
    total_values = nonzero_values.numel()
    
    # Calculate the percentage of values in the range
    percentage_in_range = (num_values_in_range / total_values) * 100
    
    return percentage_in_range.item()  # Convert to Python float

```

### mass:

```python
lower_bound = 0.75
upper_bound = 0.95

comparing_data_path = "data/dataset_distlognorm_massnorm.pt"
generated_data_path = "diffusion_notebook/T&E1000/diffusion_model_T&E1000_processed.pt"

original = torch.load(comparing_data_path)
generated = torch.load(generated_data_path, map_location = "cpu")

# Calculate the percentage of values in the range
percentage = percentage_values_in_range(original[:, 1], lower_bound, upper_bound)

print("Percentage of values between {} and {}: {:.2f}%".format(lower_bound, upper_bound, percentage))


percentage = percentage_values_in_range(generated[:, 1], lower_bound, upper_bound)

print("Percentage of values between {} and {}: {:.2f}%".format(lower_bound, upper_bound, percentage))
```
